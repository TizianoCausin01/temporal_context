```python
def create_RDM(data):
    RDM = np.corrcoef(data)
    rows, cols = np.tril_indices(RDM.shape[0], k=-1)
    RDM_vec = RDM[rows, cols]
    return RDM_vec
```

```python
import numpy as np
from scipy.spatial.distance import pdist, squareform
X = a3d 
F, D, T = X.shape
RDMs = np.zeros((D, D, T))  # store one RDM per timepoint

for t in range(T):
    # Get data at time t: shape (F, D)
    Xt = X[:, :, t]
    
    # Compute dissimilarity across datapoints (columns)
    # â†’ each column is a datapoint (so transpose)
    RDMs[:, :, t] = squareform(pdist(Xt.T, metric='correlation'))
```
