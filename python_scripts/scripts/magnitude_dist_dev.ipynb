```python
# Another, slower version of the numba compiled one we have in src
"""
Compute the pairwise difference of vector norms (L2 magnitudes) 
for the columns of x, then index upper triangular entries.

Parameters
----------
x : np.ndarray, shape (d, n)
    Columns are the vectors whose magnitudes we compare.

Returns
-------
gram : np.ndarray
    Vector containing the upper-triangular differences of norms.
"""
def magnitude_distance(x):
    # Compute column-wise norms
    norms = np.linalg.norm(x, axis=0)  # shape (n,)
    # Compute pairwise absolute differences
    diff_matrix = np.abs(norms[:, None] - norms[None, :])  # shape (n, n)
    
    # Extract upper-triangular entries using your index_gram function
    gram = index_gram(diff_matrix)
    
    return gram
```
