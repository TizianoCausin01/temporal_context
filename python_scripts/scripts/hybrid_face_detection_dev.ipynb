```python
import os, yaml, sys
import cv2
import numpy as np
from retinaface import RetinaFace
import matplotlib.pyplot as plt
%matplotlib inline

ENV = os.getenv("MY_ENV", "dev")
with open("../../config.yaml", "r") as f:
    config = yaml.safe_load(f)
paths = config[ENV]["paths"]
sys.path.append(paths["src_path"])
from image_processing.utils import read_video
```

```python

v = read_video(paths, 1, "red_to_anna2_10s.mp4")
```

```python
for i in range(10): #v.shape[0])
    img = v[i,:,:,:].copy()
    detections = RetinaFace.detect_faces(img)
    if detections:
        # Create a blank mask
        while True:
            # Loop over all detected faces
            for key in detections.keys():
                face = detections[key]
                facial_area = face["facial_area"]  # [x1, y1, x2, y2]
                x1, y1, x2, y2 = facial_area
                img_rect = img.copy()
                cv2.rectangle(img_rect, (x1, y1), (x2, y2), (0, 0, 255), 2)
            cv2.imshow("Image with rectangle", img_rect)
            key = cv2.waitKey(0) & 0xFF  
            cv2.destroyAllWindows()
            if key == ord('o'):
                print("ok!")
                break
            else:
                 ROI_manual_selection(v[i,:,:,:])
                 break
    else: 
        #mask = np.zeros(v[0,:,:,:].shape[:2], dtype=np.uint8)
        ROI_manual_selection(v[i,:,:,:])
```

```python
def ROI_manual_selection(img):
    while True:
        r = cv2.selectROI("Select area", img)
        print("ROI:", r)  # (x, y, width, height)
        x, y, w, h = map(int, r)
        img_rect = img.copy()
        cv2.rectangle(img_rect, (x, y), (x + w, y + h), (0, 0, 255), 2)
        cv2.imshow("Image with rectangle", img_rect)
        key = cv2.waitKey(0) & 0xFF  
        #ok = input("press o if it's ok")
        cv2.destroyAllWindows()
        if key == ord('o'):
            print("ok!")
            break
    return x, y, w, h
```
